```python
import pandas as pd
import matplotlib.pyplot as plt


df1 = pd.read_excel("datasets\Aracaju.xlsx")
df2 = pd.read_excel("datasets\Fortaleza.xlsx")
df3 = pd.read_excel("datasets\cNatal.xlsx")
df4 = pd.read_excel("datasets\Recife.xlsx")
df5 = pd.read_excel("datasets\Salvador.xlsx")

df = pd.concat([df1,df2,df3,df4,df5])

df['Vendas'].fillna(df["Vendas"].mean(), inplace=True)

df['Receita'] = df['Vendas'].mul(df['Qtde'])

```

```python
df["LojaID"].value_counts(ascending=False).plot.bar()
```

Faz a contagem de quantas linhas contém o id de uma loja e cria um gráfico

```python
df["LojaID"].value_counts(ascending=True).plot.barh();
```

Ordena com base no ano e faz a soma de todas as receitas referente a aquele ano

```python
df.groupby(df["Data"].dt.year)["Receita"].sum().plot.pie();
```

Ordena com base na cidade e conta quantas vendas cada cidade fez


```python
plt.style.use("ggplot") #Altera o estilo do gráfico
df["Cidade"].value_counts().plot(title = "Total de vendas por cidade", color="Red")
plt.xlabel("Cidade")
plt.ylabel("Quantidade de Vendas")
```
